**1.1 Write a Python Program(with class concepts) to find the area of the triangle using the below
formula.
 area = (s*(s-a)*(s-b)*(s-c)) ** 0.5**
 
**Function to take the length of the sides of triangle from user should be defined in the parent
class and function to calculate the area should be defined in subclass.**

```python
class triangle:
    
    def __init__(self,side1,side2,side3):
        self.a=float(side1)
        self.b=float(side2)
        self.c=float(side3)
        
class area(triangle):
    def __init__(self,*args):
        super().__init__(*args)

    def get_area(self):
        s=(self.a+self.b+self.c)/2
        return (s*(s-self.a)*(s-self.b)*(s-self.c))**0.5
```

```python
abc=area(4,4,4)
```

```python
abc.get_area()
```

# v1.2 Write a function filter_long_words() that takes a list of words and an integer n and returns the list of words that are longer than n.

```python
def filter_long_words(val,n):
    return [i for i in val if len(i)>n]

    
```

```python
l=['versha','anam','neetika','tanya','prageesha','shristhi','uma','charu']
```

```python
filter_long_words(l,5)
```

# 2.1 Write a Python program using function concept that maps list of words into a list of integers representing the lengths of the corresponding words.

```python
def output_length(val):
    return [len(i) for i in val]
```

```python
output_length(l)
```
